```python
import pandas as pd

# بيانات الطلاب
data = {
    "Name": ["Ali", "Sara", "Mona", "Omar", "Youssef"],
    "Math": [85, 90, 78, 92, 70],
    "Science": [88, 95, 80, 85, 75]
}

# نحول البيانات لجدول
df = pd.DataFrame(data)

# نحسب المتوسط
df["Average"] = (df["Math"] + df["Science"]) / 2

# نعرض الجدول
df
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Name"], df["Average"])
plt.title("Student Average Scores")
plt.xlabel("Students")
plt.ylabel("Average Score")
plt.show()
```
